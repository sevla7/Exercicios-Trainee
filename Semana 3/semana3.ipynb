# 📊 **Análise Estatística sobre o Impacto das Redes Sociais na Saúde Mental dos Estudantes**

As colunas analisadas serão:
- **Avg_Daily_Usage_Hours**: Média de horas diária na rede social
- **Study_Hours**: Horas de estudo
#
Link do dataset escolihdo: https://www.kaggle.com/datasets/shivasingh4945/student-social-media-and-mental-health-impact

```python
import pandas as pd
import scipy.stats as stats
import numpy as np
dataset = pd.read_csv('Student Social Media ANd Mental Health Impact.csv')

display(dataset)
```

### **Métricas Descritivas**

```python
coluna_estudo = dataset['Study_Hours']
coluna_horas_social = dataset['Avg_Daily_Usage_Hours']
```

#### **Média**

```python
media_estudo = coluna_estudo.mean()
media_horas_social = coluna_horas_social.mean()

print('Média das Colunas')
print(f'Qualidade de Estudo: {media_estudo:.2f}')
print(f'Horas na Rede Social: {media_horas_social:.2f}')
```

#### **Mediana**

```python
mediana_estudo = coluna_estudo.median()
mediana_horas_social = coluna_horas_social.median()

print('Mediana das Colunas')
print(f'Qualidade de Estudo: {mediana_estudo:.2f}')
print(f'Horas na Rede Social: {mediana_horas_social:.2f}')
```

#### **Moda**

```python
moda_estudo = coluna_estudo.mode()
moda_horas_social = coluna_horas_social.mode()

print('Moda das Colunas')
print(f'Qualidade de Estudo: {moda_estudo[0]:.2f}')
print(f'Horas na Rede Social: {moda_horas_social[0]:.2f}')
```

### **Métricas de Dispersão**

#### **Amplitude**

```python
amplit_estudo = coluna_estudo.max() - coluna_estudo.min()
amplit_horas_social = coluna_horas_social.max() - coluna_horas_social.min()

print('Amplitude das Colunas')
print(f'Qualidade de Estudo: {amplit_estudo:.2f}')
print(f'Horas na Rede Social: {amplit_horas_social:.2f}')
```

#### **Variância**


```python
var_estudo = coluna_estudo.var()
var_horas_social = coluna_horas_social.var() 

print('Variância das Colunas')
print(f'Qualidade de Estudo: {var_estudo:.2f}')
print(f'Horas na Rede Social: {var_horas_social:.2f}')
```

#### **Desvio Padrão**

```python
std_estudo = coluna_estudo.std()
std_horas_social = coluna_horas_social.std() 

print('Desvio padrão das Colunas')
print(f'Qualidade de Estudo: {std_estudo:.2f}')
print(f'Horas na Rede Social: {std_horas_social:.2f}')
```

#### **Teste de Hipótese**

**Hipótese estatística:** As redes sociais estão atrapalhando o rendimento ou tempo dedicado aos estudos.?
#
**Hipótese nula(H0):** Não há correlação entre as horas na rede social e as horas de estudo. Um não afeta o outro.
#
**Hipótese alternativa(Ha):** Há uma correlação negativa significativa entre as horas na rede social e as horas de estudo. Quanto mais tempo na rede social, menos tempo estudando.

- Como são duas variáveis numéricas contínuas (horas vs. horas), é uma análise de correlação.
- Irei usar a Correlação de Pearson($r$)
- $r = 1$: correlação positiva (linha linear positiva)
- $r = 0$: ausência de correlação
- $r = -1$: correlação negativa (linha linear negativa)
#
Força da relação (Regra geral):
- De 0,1 a 0,3: Correlação fraca.

- De 0,4 a 0,6: Correlação moderada.

- De 0,7 a 0,9: Correlação forte.

```python
# Calculando a correlação de Pearson e o P-valor
coeficiente_p, p_valor = stats.pearsonr(coluna_horas_social, coluna_estudo)

print(f"Coeficiente de Pearson (r): {coeficiente_p:.4f}")
print(f"p-valor: {p_valor:.4f}")
```

O teste de correlação de Pearson resultou em um coeficiente de $r = -0,876$. Esse valor indica uma correlação linear negativa muito forte entre as variáveis. Na prática, o resultado comprova que `o aumento do tempo diário gasto em redes sociais está fortemente associado à redução das horas dedicadas ao estudo.` Sendo um valor próximo de $-1$, o modelo demonstra que o comportamento dos estudantes é altamente consistente e previsível nesse aspecto.

## Exercício 1.1 - Criando Arrays com Diferentes Métodos
- Crie os seguintes arrays usando NumPy:

- Um array 1D com os valores [10, 20, 30, 40, 50] usando np.array()
- Um array com valores de 0 a 9 usando np.arange()
- Um array com 5 zeros usando np.zeros()

# Exercício 4 - Operações Elemento a Elemento
Dados os arrays:

- arr1 = np.array([10, 20, 30, 40])
- arr2 = np.array([2, 3, 4, 5])
#
 Realize as seguintes operações e exiba os resultados:

- Soma: arr1 + arr2
- Subtração: arr1 - arr2
- Multiplicação elemento a elemento: arr1 * arr2
- Divisão elemento a elemento: arr1 / arr2

# Exercício 5 - Reshape Básico
Dado o array arr = np.arange(12) (valores de 0 a 11):

- Remodele o array para uma matriz 3x4 usando .reshape(3, 4)
- Remodele o array para uma matriz 4x3
- Remodele o array para uma matriz 2x6
- Remodele o array para uma matriz 6x2
#
Use -1 para que o NumPy calcule automaticamente uma dimensão: .reshape(2, -1) 
#
Exiba cada resultado e verifique o shape de cada matriz criada.

```python
arr = np.arange(12)

reshape1 = arr.reshape(3,4)
reshape2 = arr.reshape(4,3)
reshape3 = arr.reshape(2,6)
reshape4 = arr.reshape(6,2)
reshape5 = arr.reshape(2,-1)

print(f'===Exibição===\n'
      f'Array 1: {reshape1}\n'
      f'Array 2:{reshape2}\n'
      f'Array 3:{reshape3}\n'
      f'Array 4:{reshape4}\n'
      f'Array 5: {reshape5}')
```
